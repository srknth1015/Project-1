```
# import the necessary packages
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```
edges = cv2.Canny(image,100,200)
```

```
cv2_imshow(edges)
```

```
#kernel = np.ones((3,3),np.float32)/25
#Vertical lines
#We are looking for vertical lines by accelerating center of the 3x3 matrix
#[-1,2,-1]
#[-1,2,-1]
#[-1,2,-1]
kernel = np.float32([[-1,2,-1],[-1,2,-1],[-1,2,-1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
#kernel = np.ones((3,3),np.float32)/25
#Horizontal lines
kernel = np.float32([[-1,-1,-1],[2,2,2],[-1,-1,-1]])

dst = cv2.filter2D(image,0,kernel)
cv2_imshow(dst)
```

```
#kernel = np.ones((3,3),np.float32)/25
#blur image
kernel = np.float32([[1/15,1/15,1/15],[1/15,1/15,1/15],[1/15,1/15,1/15]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
#kernel = np.ones((3,3),np.float32)
#45 degre from left
kernel = np.float32([[2,-1,-1],[-1,2,-1],[-1,-1,2]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
#kernel = np.ones((3,3),np.float32)
#Right angled 45degre
kernel = np.float32([[-1,-1,2],[-1,2,-1],[2,-1,-1]])

dst2 = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst2)
```

```
#Adding both right and left angled 45 degree
#We get a sum of both images thats capturing both the angles
m3 = dst+dst2
cv2_imshow(m3)
```

```
#kernel = np.ones((3,3),np.float32)
#image sharpanning
kernel = np.float32([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
#kernel = np.ones((3,3),np.float32)
#identity kernal
kernel = np.float32([[0,0,0],[0,1,0],[0,0,0]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```
